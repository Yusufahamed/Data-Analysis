# **Variables in Python Programming**

Variables are like container which can store the data values

```python
#num1 is called as variables.
#200 is called as values.
num1 = 200
print(num1)
```

```python
name = "Yusuf" #name is a variable and Yusuf is a value inside the variable.
print(name)
```

```python
#Longform
a=10
b=20
c=30
print(a)
print(b)
print(c)
```

```python
#shortform in one line
a, b, c = 10, 20, 30
print(a, b, c)
```

```python
# Can't start variables with numbers.
# Can't use any special characters in variables expect _ this symbol.
#1name = "yusuf" [This is a syntax error because of numerical term starts first].
#n@me = "yusuf" [This is a syntax error because every special characters are gives syntax err expect symbol_].
_name = "yusuf"
print(_name)
```

# Datatypes in Python Programming

The kind of values we actually syore in our variables are called as data types.

```python
#1) int means value of output will be in number which is not with decimal format or a number inside the comma.
num1 = 100
num2 = -100
print(num1,num2)
```

```python
#checking the type of num1 & num2
print(type(num1))
print(type(num2))
```

```python
#2) float will gives the value of output will be with decimal numers
num3 = 100.57
num4 = -200.00
print(num3,num4)
```

```python
#verify the type of data in num3 & num4
print(type(num3))
print(type(num4))
```

```python
#3) string means the value of output will be in text with the open quotation of (' ' ; " ")
name1 = "Yusuf"
subject= "PGP in Data Science&AI"
print(name1)
print(subject)
```

```python
#verify the type of data in name1 & subject
print(type(name1))
print(type(subject))
```

```python
#4) boolean means the value of output will be True or False.
a = True
b = False
print(a,b)
```

```python
#verify the type of data in a & b
print(type(a))
print(type(b))
```

```python
#5) complex means the value of output will be the combination of real(i) and imaginary(j).
com1 = 100+6j
print(com1)
print(com1.real)
print(com1.imag)
```

```python
#verify the type of data in com1
print(type(com1))
```
